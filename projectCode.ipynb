```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\sijan\Downloads\Project PC Mam\New project\laptop_price - dataset.csv")
import numpy as np
import matplotlib.pyplot as plt

# Apply log transformation
# log_prices = np.log(df['Price (Euro)'])
log_prices = (df['Price (Euro)'])
sorted_log_prices = log_prices.sort_values().reset_index(drop=True)

plt.figure(figsize=(12, 5))
plt.bar(sorted_log_prices.index, sorted_log_prices, color='orange')
plt.title("Price (Euro) Distribution")
plt.xlabel("Index (Sorted Laptops)")
plt.ylabel("Log(Price)")
plt.show()
```

```python
skew_value = df['Price (Euro)'].skew()
print("Skewness: ", skew_value)
```

```python
plt.hist(sorted_log_prices, bins=30)
plt.xlabel("Price (€)")
plt.ylabel("Frequency")
plt.title("Laptop Price Distribution")
plt.show()
```

```python
average1 = df['Inches'].mean()
average2 = df['CPU_Frequency (GHz)'].mean()
average3 = df['Weight (kg)'].mean()
average4 = df['Price (Euro)'].mean()

print(round(average1,3))
print(round(average2,3))
print(round(average3,3))
print(round(average4,3))
```

```python
a1 = df['Inches'].median()
a2 = df['CPU_Frequency (GHz)'].median()
a3 = df['Weight (kg)'].median()
a4 = df['Price (Euro)'].median()

print(round(a1,3))
print(round(a2,3))
print(round(a3,3))
print(round(a4,3))
```

```python
a1 = df['Inches'].var()
a2 = df['CPU_Frequency (GHz)'].var()
a3 = df['Weight (kg)'].var()
a4 = df['Price (Euro)'].var()

print(round(a1,3))
print(round(a2,3))
print(round(a3,3))
print(round(a4,3))
```

```python
a1 = df['Inches'].std()
a2 = df['CPU_Frequency (GHz)'].std()
a3 = df['Weight (kg)'].std()
a4 = df['Price (Euro)'].std()

print(round(a1,3))
print(round(a2,3))
print(round(a3,3))
print(round(a4,3))
```

```python

# 1️⃣ Convert 'Memory' to numeric (extract digits only)
df['Memory_Int'] = df['Memory'].str.extract(r'(\d+)').astype(int)

# 2️⃣ Frequency & Percentage for 'RAM (GB)'
ram_freq = df['RAM (GB)'].value_counts()
ram_percent = df['RAM (GB)'].value_counts(normalize=True) * 100

print("RAM (GB) Frequency:\n", ram_freq)
print("\nRAM (GB) Percentage (%):\n", ram_percent.round(2))

# 3️⃣ Frequency & Percentage for 'Memory'
mem_freq = df['Memory_Int'].value_counts()
mem_percent = df['Memory_Int'].value_counts(normalize=True) * 100

print("\nMemory Frequency:\n", mem_freq)
print("\nMemory Percentage (%):\n", mem_percent.round(2))
```

```python
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.violinplot(x='CPU_Company', y='Price (Euro)', data=df)
plt.title("Price Distribution by CPU Company")
plt.xlabel("CPU Company")
plt.ylabel("Price (Euro)")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.boxplot(x='TypeName', y='Price (Euro)', data=df, showmeans=True)
plt.title("Price Distribution by Laptop Type")
plt.xlabel("Laptop Type (TypeName)")
plt.ylabel("Price (Euro)")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count of each TypeName per Company
type_counts = df.groupby(['Company', 'TypeName']).size().unstack(fill_value=0)

# Plot as grouped bar chart
type_counts.plot(kind='bar', stacked=False, figsize=(12, 6))
plt.title("Count of Laptop Types per Company")
plt.xlabel("Company")
plt.ylabel("Number of Laptops")
plt.xticks(rotation=45)
plt.legend(title="TypeName")
plt.tight_layout()
plt.show()

```

```python
import seaborn as sns

# Select only numerical columns for correlation
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Compute correlation matrix
corr = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.tight_layout()
plt.show()
```
